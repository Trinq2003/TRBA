# II. Mark and punctuation processing

```python
import csv

def filter_rows(input_file, output_file):
    filtered_rows = []

    with open(input_file, 'r') as file:
        reader = csv.DictReader(file, delimiter='\t')
        for row in reader:
            label = row['Label']
            if not any(c.isalpha() for c in label):
                filtered_rows.append(row)

    if filtered_rows:
        fieldnames = filtered_rows[0].keys()

        with open(output_file, 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames, delimiter='\t')
            writer.writeheader()
            writer.writerows(filtered_rows)

        print(f"Filtered rows are written to {output_file}.")
    else:
        print("No rows found with label containing only marks.")

# Usage example
input_file = '../label.csv'
output_file = '../mark_only.csv'
filter_rows(input_file, output_file)
```

```python
import csv
import os

def filter_rows(input_file, output_file, image_folder):
    filtered_rows = []

    with open(input_file, 'r') as file:
        reader = csv.DictReader(file, delimiter='\t')
        for row in reader:
            label = row['Label']
            if not any(c.isalpha() for c in label):
                filtered_rows.append(row)
                image_name = row['Image']
                image_path = os.path.join(image_folder, image_name)
                if os.path.exists(image_path):
                    os.remove(image_path)
                    print(f"Removed {image_path}")

    if filtered_rows:
        fieldnames = filtered_rows[0].keys()

        with open(output_file, 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames, delimiter='\t')
            writer.writeheader()
            writer.writerows(filtered_rows)

        print(f"Filtered rows are written to {output_file}.")
    else:
        print("No rows found with label containing only marks.")

# Usage example
input_file = '../label.csv'
output_file = '../mark_only.csv'
image_folder = '/home/wallace/Code/HUST/data_alphabet'
filter_rows(input_file, output_file, image_folder)
```
